```python
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris()

Data = iris.data
Species = iris.target

print(Data)
```

```python
#task1: Cross validation
#splitting data to training and testing datasets
```

```python
#X0 = setosa
testing_X0 = Data[:10]
training_X0 = Data[10:50]

#X1 = versicolor
testing_X1 = Data[50:60]
training_X1 = Data[60:100]

#X2 = virginica
testing_X2 = Data[100:110]
training_X2 = Data[110:150]

testing_X = np.append(testing_X0,testing_X1, axis = 0)
testing_X = np.append(testing_X,testing_X2, axis = 0)

training_X = np.append(training_X0,training_X1, axis = 0)
training_X = np.append(training_X,training_X2, axis = 0)

Species_test = []

Species_test.append(Species[:10])
Species_test.append(Species[50:60])
Species_test.append(Species[-10:])

print(training_X)
print(testing_X)
```

```python
#Functions for mean and st_dev
```

```python
from math import sqrt
from math import exp
from math import pi

# Calculate the mean 
def mean(values):
    return sum(values)/float(len(values))

# Calculate the standard deviation 
def st_deviation(values):
    average = mean(values)
    return sqrt(sum([(x-average)**2 for x in values]) / float(len(values)-1))
```

```python
#Setosa mean
mean_X0 = []
feature = []
for i in range(4):
  for j in range(40):
    feature.append((training_X0[j,i]))
  mean_X0.append(mean(feature))
  feature = []

print(mean_X0)

#Versicolor mean
mean_X1 = []
feature = []
for i in range(4):
  for j in range(40):
    feature.append((training_X1[j,i]))
  mean_X1.append(mean(feature))
  feature = []

print(mean_X1)

#Virginica mean
mean_X2 = []
feature = []
for i in range(4):
  for j in range(40):
    feature.append((training_X2[j,i]))
  mean_X2.append(mean(feature))
  feature = []

print(mean_X2)
```

```python
#setosa_st_deviation

feature = []
st_dev_X0 = []
for i in range(4):
  for j in range(40):
    feature.append(training_X0[j,i])
  st_dev_X0.append(st_deviation(feature))
  feature = []

print(st_dev_X0)

#Versicolor_st_deviation

feature = []
st_dev_X1 = []
for i in range(4):
  for j in range(40):
    feature.append(training_X1[j,i])
  st_dev_X1.append(st_deviation(feature))
  feature = []

print(st_dev_X1)

#Virginica_st_deviation

feature = []
st_dev_X2 = []
for i in range(4):
  for j in range(40):
    feature.append(training_X2[j,i])
  st_dev_X2.append(st_deviation(feature))
  feature = []

print(st_dev_X2)
```

```python
#Task2: Likelihood
```

```python
def like_fun(x, mean, st_dev):
    denominator = 2 * float(st_dev) ** 2
    number = (2 * pi ** 0.5) * denominator
    main_func = exp(-(float(x)-float(mean)) ** 2 / denominator)
    return main_func / number
Likelihood = []
feature = []
values_X0 = []
values_X1 = []
values_X2 = []
for i in range(30):
  for j in range(4):
    number = testing_X[i,j]
    values_X0.append(like_fun(number,mean_X0[j],st_dev_X0[j]))
    values_X1.append(like_fun(number,mean_X1[j],st_dev_X1[j]))
    values_X2.append(like_fun(number,mean_X2[j],st_dev_X2[j]))
  feature.append(values_X0)
  feature.append(values_X1)
  feature.append(values_X2)
  Likelihood.append(feature)
  feature = []
  values_X0 = []
  values_X1 = []
  values_X2 = []

for i in range(30):
  print(Likelihood[i])
```

```python
#Task3: Priori
```

```python
f = 2
Priori_X0=len(training_X0)/len(training_X)
Priori_X1=len(training_X1)/len(training_X)
Priori_X2=len(training_X2)/len(training_X)
print("{:.{}f}".format(Priori_X0, f))
print("{:.{}f}".format(Priori_X1, f))
print("{:.{}f}".format(Priori_X2, f))
```

```python
#Task4: Posterior
```

```python
from numpy.ma.core import argmax
from numpy import prod
posterior = []
feature = []
for i in range(30):
  for j in range(3):
    post = (prod(Likelihood[i][j]) * Priori_X0)
    feature.append(post)
    post = 0
  posterior.append(feature)
  feature = []

for j in range(30):
  print(argmax(posterior[j]))
```

```python
#Task5: Accuracy
```

```python
accuracy = []
count = 0
for i in range(3):
  for j in range(10):
    accuracy.append(Species_test[i][j])
for i in range(30):
  if(argmax(posterior[i]) == accuracy[i]):
    count += 1
print(count / len(testing_X) * 100)
```
